```python
emp
```

```python
emp['Name']
```

```python
emp['Name'] = emp['Name'].str.replace(r'\W','',regex = True)              # W is used \W matches any character that is NOT a word character. remove regular expression like ^,*,
emp
```

```python
emp['Domain'] = emp['Domain'].str.replace(r'\W','',regex = True)
emp
```

```python
emp['Age'] = emp['Age'].str.replace(r'\W','',regex = True)
emp
```

```python
emp['Location'] = emp['Location'].str.replace(r'\W','',regex = True)
emp
```

```python
emp['Age']
```

```python
emp['Age'] = emp['Age'].str.extract('(\\d+)')     #  d+ “Extract one or more digits from the string.”  ex: 20 yrs --> 20
```

```python
emp['Age']
```

```python
emp['Salary'] = emp['Salary'].str.replace(r'\W','',regex = True)       #clean salary column
emp
```

```python
emp['Salary']
```

```python
emp['Exp'] = emp['Exp'].str.extract('(\\d+)')
emp
```
