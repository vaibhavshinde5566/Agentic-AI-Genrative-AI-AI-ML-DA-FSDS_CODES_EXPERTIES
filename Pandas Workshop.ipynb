# Bivariate analysis (plot the graph using 2 variable)

```python
# box plot
vis3 = sns.boxplot(data = df,x = 'IncomeGroup', y = 'BirthRate')
```

```python
vis4 = sns.lmplot(data = df,x = 'InternetUsers',y = 'BirthRate')
```

```python
vis5 = sns.lmplot(data = df,x = 'InternetUsers', y = 'BirthRate',fit_reg = False)
```

```python
vis6 = sns.lmplot(data=df, x= 'InternetUsers', y ='BirthRate',fit_reg=True,hue='IncomeGroup') #hue gives color information
plt.show()
```
