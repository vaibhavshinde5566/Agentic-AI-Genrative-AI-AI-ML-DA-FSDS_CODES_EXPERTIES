# Formal Argument And Actual Argument

```python
def person(name,age):        # formal argument
    print(name)
    print(age)
person('vaibhav',22)        # Actual argument (positional argument , keyword argument , default)
```

```python
def person(name,age):       # first arg = 23 so name = 23 and age = vaibhav
    print(name)
    print(age)
    
person(23,'vaibhav')          
```

### positional argument

```python
#positional argument
def person(name,age):
    print(name)
    print(age)
    
person('vaibhav',23,34)      
```

```python
def person(name,age,city):             #defined 3 arg so passed in func also 3
    print(name)
    print(age)
    print(city)
    
person('vaibhav',23,'hyd')
```

### Keyword argument

```python
# Keyword argument
def person(name,age):
    print(name)
    print(age)
    
person(age = 22,name = 'vaibhav')          # keyword argument
```

```python
#positional argument
def person(name,age):
    print(name)
    print(age)
    
person(name = 'vaibhav',age = 22) 
```

```python
def person(name,age,city):
    print(name)
    print(age+1)
    print(city)
    
person(age=23, name='vaibhav', city = 'hyd')
```

```python
#cheack even odd 
x = int(input('Enter a number: '))
def even(x):
    if x%2==0:
        print('even number')
    else:
        print('odd number')
even(x)
```

```python
# print table using function
def table(x):
    for i in range(1,11):
        c = i*x;
        print(c)
        
table(2)
```

# Recursion

```python
def fact(n):
    f=1
    for i in range(1,n+1):
        f = f*i
        return f
fact(6)
        
```

```python
# Factorial using recursion
def fact(n):
    if n==0:
        return 1
    return n * fact(n-1)
result = fact(5)
result
```
